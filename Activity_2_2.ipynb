**Discussion:**

In Python, variables are used to store data, acting as named references to memory locations where values can be assigned and modified. These values can be of various data types, such as integers, strings, floats, or lists.

 Constants are variables whose values should not change during the program's execution. Although Python does not have built-in support for constants, you can use a naming convention to indicate that a variable should be treated as a constant.

Literals are notations for representing fixed values in source code. They are constants that are self-explanatory and do not need to be computed or evaluated. Python supports various types of literals, such as numeric literals, string literals, Boolean literals, and more.

```python
# Place all the output of the procedures here

import sys

# Function to get a grade input from the user and validate it
def get_grade(label):
    grade = float(input(f"Enter {label} Grade: "))
    if grade > 100:
        print("Invalid Grade")
        sys.exit()
    return grade     #return if the grade input exceeded the given range

# Function to calculate class standing based on weighted percentages
def calculate_class_standing(HOA, QUIZ, ASSIGN):
    return HOA * .5 + QUIZ * .3 + ASSIGN * .2

# Function to calculate final grade based on class standing and exam
def calculate_final(cs, exam):
    return cs * .5 + exam * .5

# Get student's name
name = input("Enter your Name: ")
print("\n")

# PRELIM GRADE
print("PRELIM GRADE")
PHOA = get_grade("Hands-on Activity")
PQUIZ = get_grade("Quiz")
ASSIGN = get_grade("Assignment")
PCS = calculate_class_standing(PHOA, PQUIZ, ASSIGN)
print("Prelim Class Standing is:", PCS)

PE = get_grade("Prelim Exam")
PG = calculate_final(PCS, PE)
print("Prelim Grade is:", PG)
print("\n")

# MIDTERM GRADE
print("MIDTERM GRADE")
MHOA = get_grade("Midterm Hands-on Activity")
MQUIZ = get_grade("Midterm Quiz")
MASSIGN = get_grade("Midterm Assignment")
MCS = calculate_class_standing(MHOA, MQUIZ, MASSIGN)
print("Midterm Class Standing is:", MCS)

ME = get_grade("Midterm Exam")
MG = calculate_final(MCS, ME)
print("Midterm Grade is:", MG)
print("\n")

# FINAL GRADE
print("FINAL GRADE")
FHOA = get_grade("Final Hands-on Activity")
FQUIZ = get_grade("Final Quiz")
FASSIGN = get_grade("Final Assignment")
FCS = calculate_class_standing(FHOA, FQUIZ, FASSIGN)
print("Final Class Standing9 is:", FCS)

FE = get_grade("Final Exam")
FG = calculate_final(FCS, FE)
print("Final Grade is:", FG)
print("\n")

# SUMMARY OF GRADES
print("SUMMARY OF GRADES")
print("Name:", name)
print("Prelim Grade:", PG)
print("Midterm Grade:", MG)
print("Final Grade:", FG)


#CONVERSION OF FINAL GRADES
print (f"Conversion of Final Grade to UCC's numerical Grading System")
if FG >= 99 < 100:
  print ("Your Final Grade is 1.00")
elif FG >= 95 < 98:
  print ("Your Final Grade is 1.25")
elif FG >= 90 < 94:
  print ("Your Final Grade is 1.50")
elif FG >= 85 < 89:
  print ("Your Final Grade is 1.75")
elif FG >= 80 < 84:
  print ("Your Final Grade is 2.00")
elif FG >= 75 < 79:
  print ("Your Final Grade is 2.25")
elif FG >= 70 < 74:
  print ("Your Final Grade is 2.50")
elif FG >= 65 < 69:
  print ("Your Final Grade is 2.75")
elif FG >= 60 < 64:
  print ("Your Final Grade is 3.00")
else:
  print ("Failed")
```
